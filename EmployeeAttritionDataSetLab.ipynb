This code will import the necessary libraries, matplotlib.pylot and seaborn, to enable plotting the attrition rates for categorical features. Then, I will start calculating the attrition rates for the 'Department' feature and visualize it.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

print("Attrition Rates by Department\n")

# Calculate attrition rate for each department
department_attrition = df.groupby(['Department', 'Attrition']).size().unstack(fill_value=0)
department_attrition['Total'] = department_attrition.sum(axis=1)
department_attrition['AttritionRate'] = (department_attrition['Yes'] / department_attrition['Total']) * 100

print(department_attrition[['Yes', 'No', 'AttritionRate']])

# Plotting attrition rate by Department
plt.figure(figsize=(10, 6))
sns.barplot(x=department_attrition.index, y=department_attrition['AttritionRate'], palette='viridis')
plt.title('Attrition Rate by Department', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Attrition Rate (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Now that the data availability has been confirmed, I will proceed with hyperparameter tuning for the Decision Tree Classifier using GridSearchCV, defining a parameter grid for max_depth, min_samples_leaf, and criterion to find the optimal combination.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Define the parameter grid for GridSearchCV
param_grid = {
    'max_depth': [3, 5, 7, 10, None],  # None means no limit on depth
    'min_samples_leaf': [1, 5, 10, 20],
    'criterion': ['gini', 'entropy']
}

# Initialize a Decision Tree Classifier (base estimator for GridSearchCV)
dt = DecisionTreeClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=1)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

print("GridSearchCV complete.")
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best F1 score: {grid_search.best_score_:.4f}")
```

Now that the optimal hyperparameters have been identified, I will train a new Decision Tree Classifier using these best parameters. After training, I will make predictions on the test set and evaluate the model's performance using various metrics (accuracy, precision, recall, F1-score) and visualize the confusion matrix.


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Get the best parameters from GridSearchCV
best_params = grid_search.best_params_

# Initialize a new Decision Tree Classifier with the optimal hyperparameters
dt_tuned = DecisionTreeClassifier(random_state=42, **best_params)

# Train the tuned Decision Tree Classifier using the training data
dt_tuned.fit(X_train, y_train)

# Make predictions on the test set with the tuned model
y_pred_tuned = dt_tuned.predict(X_test)

# Calculate evaluation metrics for the tuned model
accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
precision_tuned = precision_score(y_test, y_pred_tuned)
recall_tuned = recall_score(y_test, y_pred_tuned)
f1_tuned = f1_score(y_test, y_pred_tuned)

print("### Tuned Decision Tree Classifier Performance\n")
print(f"Best Hyperparameters: {best_params}")
print(f"Accuracy: {accuracy_tuned:.4f}")
print(f"Precision: {precision_tuned:.4f}")
print(f"Recall: {recall_tuned:.4f}")
print(f"F1-Score: {f1_tuned:.4f}")

# Generate confusion matrix for the tuned model
cm_tuned = confusion_matrix(y_test, y_pred_tuned)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_tuned, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['No Attrition', 'Yes Attrition'], yticklabels=['No Attrition', 'Yes Attrition'])
plt.title('Confusion Matrix for Tuned Decision Tree Classifier', fontsize=16)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.show()
```
